```python
import pandas as pd
df=pd.read_csv('customer_shopping_behaviour.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
missing_cols=df.isnull().sum()
print(missing_cols)
```

```python
median_ratings=df.groupby('Category')['Review Rating'].transform('median')
df['Review Rating']=df['Review Rating'].fillna(median_ratings)
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
print(df.columns)
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
print(df.columns)
```

```python
labels=['Young','Adult','Middle_Age','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
df[['age','age_group']].head()
```

```python
df['frequency_of_purchases'].unique()
```

```python
freq_map={
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Bi-Weekly': 14,
    'Quarterly': 90,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(freq_map)
df[['purchase_frequency_days','frequency_of_purchases']].head()
```

```python
df[['discount_applied','promo_code_used']].head()
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df=df.drop('promo_code_used',axis=1)
```
